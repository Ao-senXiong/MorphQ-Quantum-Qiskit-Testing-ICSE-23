```python
from torch_two_sample import statistics_diff
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SAMPLE_SIZE = 100
DIM = 20
n1 = SAMPLE_SIZE
n2 = SAMPLE_SIZE


d1 = np.around(np.random.rand(n1, DIM))
d2 = np.around(np.random.rand(n2, DIM))
```

```python
plt.subplot(121)
plt.imshow(d1, cmap='Greys_r', interpolation='nearest')
plt.title("D1")
plt.subplot(122)
plt.imshow(d2, cmap='Greys_r', interpolation='nearest')
plt.title("D2")
plt.tight_layout()
plt.show()
```

```python
energy_test = statistics_diff.EnergyStatistic(n1, n2)
```

```python
import torch
sample_1 = torch.from_numpy(d1)
sample_2 = torch.from_numpy(d2)

statistic, dist_matrix = energy_test.__call__(sample_1, sample_2, ret_matrix=True)
dist_matrix = dist_matrix.to(torch.float32)
p_value = energy_test.pval(dist_matrix, n_permutations=1000)
```
